### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python

loja.info()
```

```python
Total_Faturamento_loja_1 = loja.Preço.sum().round(2)
Total_Faturamento_loja_2 = loja2.Preço.sum().round(2)
Total_Faturamento_loja_3 = loja3.Preço.sum().round(2)
Total_Faturamento_loja_4 = loja4.Preço.sum().round(2)
```

```python
print(f'O faturamento da loja 1: R$ {Total_Faturamento_loja_1} ')
print(f'O faturamento da loja 2: R$ {Total_Faturamento_loja_2} ')
print(f'O faturamento da loja 3: R$ {Total_Faturamento_loja_3} ')
print(f'O faturamento da loja 4: R$ {Total_Faturamento_loja_4} ')
```

# 2. Vendas por Categoria

```python
vendas_categoria_loja_1 = loja.groupby('Categoria do Produto')[['Preço']].sum()
vendas_categoria_loja_2 = loja2.groupby('Categoria do Produto')[['Preço']].sum()
vendas_categoria_loja_3 = loja3.groupby('Categoria do Produto')[['Preço']].sum()
vendas_categoria_loja_4 = loja4.groupby('Categoria do Produto')[['Preço']].sum()
```

```python
print(f'\nVenda por Categoria loja 1: {vendas_categoria_loja_1}')
print(f'\nVenda por Categoria loja 2: {vendas_categoria_loja_2}')
print(f'\nVenda por Categoria loja 3: {vendas_categoria_loja_3}')
print(f'\nVenda por Categoria loja 4: {vendas_categoria_loja_4}')
```

# 3. Média de Avaliação das Lojas


```python
avaliação_media_loja_1 = loja['Avaliação da compra'].mean().round(2)
avaliação_media_loja_2 = loja2['Avaliação da compra'].mean().round(2)
avaliação_media_loja_3 = loja3['Avaliação da compra'].mean().round(2)
avaliação_media_loja_4 = loja4['Avaliação da compra'].mean().round(2)
```

```python
print(f'Avaliação da compra loja 1: {avaliação_media_loja_1}')
print(f'Avaliação da compra loja 2: {avaliação_media_loja_2}')
print(f'Avaliação da compra loja 3: {avaliação_media_loja_3}')
print(f'Avaliação da compra loja 4: {avaliação_media_loja_4}')
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_mais_vendidos_loja_1 = loja['Produto'].value_counts()
produtos_mais_vendidos_loja_2 = loja2['Produto'].value_counts()
produtos_mais_vendidos_loja_3 = loja3['Produto'].value_counts()
produtos_mais_vendidos_loja_4 = loja4['Produto'].value_counts()
```

```python
print(f'\nProdutos mais vendidos loja_1:\n{produtos_mais_vendidos_loja_1.head(3)}')
print(f'\nProdutos mais vendidos loja_2:\n{produtos_mais_vendidos_loja_2.head(3)}')
print(f'\nProdutos mais vendidos loja_3:\n{produtos_mais_vendidos_loja_3.head(3)}')
print(f'\nProdutos mais vendidos loja_4:\n{produtos_mais_vendidos_loja_4.head(3)}')
```

```python
produtos_menos_vendidos_loja_1 = loja['Produto'].value_counts(ascending=True)
produtos_menos_vendidos_loja_2 = loja2['Produto'].value_counts(ascending=True)
produtos_menos_vendidos_loja_3 = loja3['Produto'].value_counts(ascending=True)
produtos_menos_vendidos_loja_4 = loja4['Produto'].value_counts(ascending=True)
```

```python
print(f'\nProdutos menos vendidos loja_1:{produtos_menos_vendidos_loja_1.head(3)}')
print(f'\nProdutos menos vendidos loja 2:{produtos_menos_vendidos_loja_2.head(3)}')
print(f'\nProdutos menos vendidos loja 3:{produtos_menos_vendidos_loja_3.head(3)}')
print(f'\nProdutos menos vendidos loja 4:{produtos_menos_vendidos_loja_4.head(3)}')
```

```python
lojas = [
    produtos_menos_vendidos_loja_1,
    produtos_menos_vendidos_loja_2,
    produtos_menos_vendidos_loja_3,
    produtos_menos_vendidos_loja_4
]

for i, produtos in enumerate(lojas, start=1):
    print(f'\nProdutos menos vendidos loja {i}:')
    print(produtos.head(3))
```

# 5. Frete Médio por Loja

```python
frete_medio_loja_1 = loja['Frete'].mean().round(2)
frete_medio_loja_2 = loja2['Frete'].mean().round(2)
frete_medio_loja_3 = loja3['Frete'].mean().round(2)
frete_medio_loja_4 = loja4['Frete'].mean().round(2)
```

```python
print(f'Frete médio loja 1: R$ {frete_medio_loja_1}')
print(f'Frete médio loja 2: R$ {frete_medio_loja_2}')
print(f'Frete médio loja 3: R$ {frete_medio_loja_3}')
print(f'Frete médio loja 4: R$ {frete_medio_loja_4}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['Loja1', 'Loja2', 'Loja3', 'Loja4']
Total_Faturamento_lojas = [Total_Faturamento_loja_1, Total_Faturamento_loja_2, Total_Faturamento_loja_3, Total_Faturamento_loja_4]

df = pd.DataFrame({'Loja': lojas, 'Faturamento': Total_Faturamento_lojas})
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.bar(df['Loja'], df['Faturamento'], color='blue') # Cria o gráfico de barras
plt.xlabel('Lojas')  # Define o rótulo do eixo x
plt.ylabel('Faturamento Total (R$)')  # Define o rótulo do eixo y
plt.title('Faturamento Total por Loja')  # Define o título do gráfico
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor visualização
plt.tight_layout()  # Ajusta o layout para evitar cortes

for i in range(len(lojas)):
      plt.text(i, Total_Faturamento_lojas[i] + 10000,
             f"R$ {Total_Faturamento_lojas[i]/1000000:.2f}M",
             ha='center', fontsize=11)
plt.show()  # Mostra o gráfico
```

```python
lojas = ['Loja1', 'Loja2', 'Loja3', 'Loja4']
avaliação_media_loja = [avaliação_media_loja_1, avaliação_media_loja_2, avaliação_media_loja_3, avaliação_media_loja_4]
df = pd.DataFrame({'Loja': lojas, 'Avaliação media': avaliação_media_loja})
```

```python
# Criar o gráfico de barras horizontal
plt.barh(df['Loja'], df['Avaliação media'])

# Adicionar rótulos e título
plt.xlabel('Avaliação média')
plt.ylabel('Lojas')
plt.title('Avaliação Média por Loja')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Exibir o gráfico
plt.show()
```

```python
lojas = ['Loja1', 'Loja2', 'Loja3', 'Loja4']
faturamento_medio_loja_1 = loja['Preço'].mean().round(2)
faturamento_medio_loja_2 = loja2['Preço'].mean().round(2)
faturamento_medio_loja_3 = loja3['Preço'].mean().round(2)
faturamento_medio_loja_4 = loja4['Preço'].mean().round(2)
faturamento_medio_lojas = [faturamento_medio_loja_1, faturamento_medio_loja_2, faturamento_medio_loja_3, faturamento_medio_loja_4]
df = pd.DataFrame({'Loja': lojas, 'Faturamento médio': faturamento_medio_lojas})
```

```python
plt.figure(figsize=(8,5))

# Plot the faturamento_medio using the 'df' DataFrame from the previous cell
plt.bar(df['Loja'], df['Faturamento médio'], color=['green','green','green','red'])

plt.title('Faturamento Médio por Loja')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylabel('Faturamento Médio')
plt.xlabel('Lojas')
plt.xticks(rotation=60)

for i in range(len(lojas)):
      plt.text(i, faturamento_medio_lojas[i] + 10,
             f"R$ {faturamento_medio_lojas[i]:,.2f}",
             ha='center', fontsize=11)
plt.show()
```

```python
lojas = ['Loja1', 'Loja2', 'Loja3', 'Loja4']
frete_medio = [frete_medio_loja_1, frete_medio_loja_2, frete_medio_loja_3, frete_medio_loja_4]
df = pd.DataFrame({'Loja': lojas, 'frete medio': frete_medio})
```

## Objetivo

O senhor João possui **4 lojas** e deseja vender uma delas para investir em um novo negócio.  
Para apoiá-lo na tomada de decisão, foram analisadas as seguintes métricas:

1. Faturamento total de cada loja  
2. Categorias mais populares  
3. Média de avaliação dos clientes  
4. Produtos mais e menos vendidos  
5. Custo médio do frete  

---
## **Análises**


### 1. Faturamento Total

- Loja 1: 1.534.090,12
- Loja 2: 1.489.854,06
- Loja 3: 1.464.025,03
- Loja 4: 1.384.497,58


A Loja 1 apresenta o maior faturamento, enquanto a Loja 4 possui o menor faturamento.

---

### 2. Categorias Mais Populares
- Loja 1: Eletrônicos e eletrodomésticos  
- Loja 2: Eletrônicos e eletrodomésticos  
- Loja 3: Eletrônicos e eletrodomésticos  
- Loja 4: Eletrônicos e eletrodomésticos  

Conforme a análise, as categorias mais populares de vendas em todas as lojas são eletrônicos e eletrodomésticos.

---

### 3. Média de Avaliação dos Clientes
- Loja 1: 3,98  
- Loja 2: 4,04  
- Loja 3: 4,05  
- Loja 4: 4,00  

As médias de avaliação se mantêm equilibradas entre as lojas. A Loja 1 possui a menor nota, enquanto a Loja 3 apresenta a maior média.

---

### 4. Produtos Mais e Menos Vendidos

#### Loja 1
- Mais vendidos: Micro-ondas, TV LED UHK 4K, Guarda-roupas  
- Menos vendidos: Headset, Celular ABXY, Panela de pressão  

#### Loja 2
- Mais vendidos: Iniciação em Programação, Micro-ondas, Bateria  
- Menos vendidos: Jogo de tabuleiro, Mesa de jantar, Impressora  

#### Loja 3

---

### 5. Frete Médio

- Loja 1: 34,69  
- Loja 2: 33,62

A Loja 1 possui o frete médio mais caro, enquanto a Loja 4 apresenta o frete médio mais barato.



## **CONCLUSÃO**


Com base nas análises realizadas, levando em consideração o faturamento total, avaliação de compra e valor médio do frete das lojas.

A loja 1 embora tenha tido a menor avaliação de compra, o frete médio mais caro, se destacou por ter o maior faturamento em relação a loja 4 que teve nota de avaliação maior do que a loja 1 e o frete médio mais barato, porém com o menor faturamento entre as demais lojas.

Deste forma recomenda-se que a loja 4 seja vendida.
